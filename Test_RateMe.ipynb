```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from rateme.utils import RateMe, CAM
```

## Run through test images

```python
# load net
net = RateMe()

# get all jpegs in test_imgs folder
test_dir = 'test_imgs'
imgs_names = [os.path.join(test_dir, i) for i in os.listdir(test_dir) if i.endswith('.jpg')]
imgs = [cv2.imread(i) for i in imgs_names]
results = [str(net.predict(i)) for i in imgs]
    
# plot result
# [::,::,::-1] converts from BGR (opencv uses it) to RGB format
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.imshow(imgs[0][::,::,::-1])
plt.title(results[0])
plt.subplot(132)
plt.imshow(imgs[1][::,::,::-1])
plt.title(results[1])
plt.subplot(133)
plt.imshow(imgs[2][::,::,::-1])
plt.title(results[2])
```
